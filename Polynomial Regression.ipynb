```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\tonmo\Desktop\ML and DL\Machine Learning\Datasets\Ice_cream selling data.csv")
```

```python
dataset.head(5)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
plt.scatter(dataset["Temperature (°C)"],dataset["Ice Cream Sales (units)"])
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.show()
```

```python
dataset.corr()
```

```python
x = dataset[["Temperature (°C)"]]
y = dataset[["Ice Cream Sales (units)"]]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
prd = lr.predict(x)
```

```python
plt.scatter(dataset["Temperature (°C)"],dataset["Ice Cream Sales (units)"])
plt.plot(dataset["Temperature (°C)"],prd,c='red')
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.show()
```
